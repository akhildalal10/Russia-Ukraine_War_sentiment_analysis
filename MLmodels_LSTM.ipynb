```python
import re
import string  # Import the string module

def clean_tweets(tweets):
    # Remove RT handles
    tweets = tweets.str.replace(r'RT @[\w]*:', '', regex=True)
    # Remove Twitter handles (@xxx)
    tweets = tweets.str.replace(r'@[A-Za-z0-9]+', '', regex=True)
    # Remove URL links (httpxxx)
    tweets = tweets.str.replace(r'https?://[A-Za-z0-9./]*', '', regex=True)
    # Remove email addresses
    tweets = tweets.str.replace(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', '', regex=True)
    # Remove emojis
    EMOJI_PATTERN = re.compile(
        "["
        "\U0001F1E0-\U0001F1FF"  # flags (iOS)
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F700-\U0001F77F"  # alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251" 
        "]+"
    )
    tweets = tweets.str.replace(EMOJI_PATTERN, '', regex=True)
    # Remove single and double quotes
    tweets = tweets.str.replace(r'[\'\"]', '', regex=True)

    # tweets = tweets.str.replace({'["\']': ''}, regex=True)
    # Remove single letters
    tweets = tweets.apply(lambda x: ' '.join([word for word in x.split() if len(word) > 1]))
    # Remove numbers
    tweets = tweets.str.replace(r'\d+', '', regex=True)
    # Remove ellipsis
    tweets = tweets.str.replace('…', '', regex=True)
    # Remove hashtags
    tweets = tweets.str.replace(r'#\w+', '', regex=True)
    # Remove punctuation
    tweets = tweets.str.replace('[{}]'.format(string.punctuation), '', regex=True)
    # Remove brackets, hyphens, and other specified characters
    tweets = tweets.str.replace(r'[\[\]\-]', '', regex=True)
    # Remove double and single quotes
    tweets = tweets.replace({'["\']': ''}, regex=True)
    # Convert to lowercase
    tweets = tweets.str.lower()
    # Remove extra whitespaces
    tweets = tweets.str.replace(r'\s+', ' ', regex=True)
    # Strip leading and trailing whitespaces
    tweets = tweets.str.strip()
    return tweets
```

```python
def preprocess_tweets(tweets):
    preprocessed_tweets = []
    for tweet in tweets:
        doc = nlp(tweet)
        
        preprocessed_tweet = [token.lemma_ for token in doc if not token.is_stop]
        
        preprocessed_tweets.append(preprocessed_tweet)
    
    return preprocessed_tweets
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python
def get_sentiment(tweet):
    # Analyze sentiment of the tweet
    sentiment = analyzer.polarity_scores(tweet)
    
    # Determine sentiment category
    if sentiment['compound'] >= 0.05:
        return 'Positive'
    elif sentiment['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
import numpy as np
from gensim.models import Word2Vec
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Assuming you have a DataFrame 'df' with a 'cleaned_tweets' column and a 'sentiment' column
# Replace 'cleaned_tweets' and 'sentiment' with the appropriate column names from your DataFrame

# Step 1: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['cleaned_tweets'], df['sentiment'], test_size=0.2, random_state=42)

# Step 2: Train Word2Vec model
tokenized_tweets = [tweet.split() for tweet in X_train]
word2vec_model = Word2Vec(sentences=tokenized_tweets, vector_size=100, window=5, min_count=1, workers=4)

# Step 3: Generate document vectors
def generate_doc_vector(tokens):
    vector = np.zeros(word2vec_model.vector_size)
    num_tokens = 0
    for token in tokens:
        if token in word2vec_model.wv:
            vector += word2vec_model.wv[token]
            num_tokens += 1
    if num_tokens != 0:
        vector /= num_tokens
    return vector

X_train_word2vec = np.array([generate_doc_vector(tokens) for tokens in tokenized_tweets])

# Step 4: Transform the test data to Word2Vec embeddings
tokenized_test_tweets = [tweet.split() for tweet in X_test]
X_test_word2vec = np.array([generate_doc_vector(tokens) for tokens in tokenized_test_tweets])

# Step 5: Train logistic regression model
logistic_reg_model = LogisticRegression(max_iter=1000)
logistic_reg_model.fit(X_train_word2vec, y_train)

# Step 6: Evaluate the model
y_pred = logistic_reg_model.predict(X_test_word2vec)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
#ETC
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import ExtraTreesClassifier
import numpy as np
from gensim.models import Word2Vec
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'df' with a 'cleaned_tweets' column and a 'sentiment' column
# Replace 'cleaned_tweets' and 'sentiment' with the appropriate column names from your DataFrame

# Step 1: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['cleaned_tweets'], df['sentiment'], test_size=0.2, random_state=42)

# Step 2: Transform the text data to TF-IDF vectors
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Step 3: Train ETC model on TF-IDF vectors
etc_model_tfidf = ExtraTreesClassifier()
etc_model_tfidf.fit(X_train_tfidf, y_train)

# Step 4: Evaluate the model
y_pred_tfidf = etc_model_tfidf.predict(X_test_tfidf)
print("TF-IDF Classification Report:")
print(classification_report(y_test, y_pred_tfidf))

# Step 5: Transform the text data to Bag-of-Words (BOW) representation
bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)

# Step 6: Train ETC model on BOW representation
etc_model_bow = ExtraTreesClassifier()
etc_model_bow.fit(X_train_bow, y_train)

# Step 7: Evaluate the model
y_pred_bow = etc_model_bow.predict(X_test_bow)
print("BOW Classification Report:")
print(classification_report(y_test, y_pred_bow))

# Step 8: Train Word2Vec model
tokenized_tweets = [tweet.split() for tweet in X_train]
word2vec_model = Word2Vec(sentences=tokenized_tweets, vector_size=100, window=5, min_count=1, workers=4)

# Step 9: Generate document vectors for training data
def generate_doc_vector(tokens):
    vector = np.zeros(word2vec_model.vector_size)
    num_tokens = 0
    for token in tokens:
        if token in word2vec_model.wv:
            vector += word2vec_model.wv[token]
            num_tokens += 1
    if num_tokens != 0:
        vector /= num_tokens
    return vector

X_train_word2vec = np.array([generate_doc_vector(tokens) for tokens in tokenized_tweets])

# Step 10: Transform the test data into Word2Vec embeddings
tokenized_test_tweets = [tweet.split() for tweet in X_test]
X_test_word2vec = np.array([generate_doc_vector(tokens) for tokens in tokenized_test_tweets])

# Step 11: Train ETC model on Word2Vec embeddings
etc_model_word2vec = ExtraTreesClassifier()
etc_model_word2vec.fit(X_train_word2vec, y_train)

# Step 12: Evaluate the model
y_pred_word2vec = etc_model_word2vec.predict(X_test_word2vec)
print("Word2Vec Classification Report:")
print(classification_report(y_test, y_pred_word2vec))

# Step 13: Compute confusion matrices
cm_tfidf = confusion_matrix(y_test, y_pred_tfidf)
cm_bow = confusion_matrix(y_test, y_pred_bow)
cm_word2vec = confusion_matrix(y_test, y_pred_word2vec)

# Step 14: Plot the confusion matrices as heatmaps
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.heatmap(cm_tfidf, annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title('Confusion Matrix (TF-IDF)')
axes[0].set_xlabel('Predicted Labels')
axes[0].set_ylabel('True Labels')

sns.heatmap(cm_bow, annot=True, fmt='d', cmap='Blues', ax=axes[1])
axes[1].set_title('Confusion Matrix (BOW)')
axes[1].set_xlabel('Predicted Labels')
axes[1].set_ylabel('True Labels')

sns.heatmap(cm_word2vec, annot=True, fmt='d', cmap='Blues', ax=axes[2])
axes[2].set_title('Confusion Matrix (Word2Vec)')
axes[2].set_xlabel('Predicted Labels')
axes[2].set_ylabel('True Labels')

plt.tight_layout()
plt.show()
```
